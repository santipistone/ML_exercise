```python
from sklearn import datasets
import numpy as np
from sklearn import preprocessing

data = datasets.load_wine()

X = data.data
y = data.target
np.random.shuffle(X)
np.random.shuffle(y)
X_train = X[:100]
y_train = y[:100]

X_test = X[100:]
y_test = y[100:]
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score

logistic = LogisticRegression(multi_class='multinomial', solver='saga')

logistic.fit(X_train, y_train)

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(logistic, X, y, scoring='accuracy', cv=cv, n_jobs=-1)

#y_hat = logistic.predict(X_test)

#vali_accuracy = sum(y_hat == y_test) / len(y_test)
print('Mean Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```
