```python
from sklearn import datasets
import numpy as np

data = datasets.load_wine()

X = data.data
y = data.target

np.random.shuffle(X)
np.random.shuffle(y)
print ("Original dataset : ", X[0])

X_train = X[:100]
y_train = y[:100]

X_test = X[100:]
y_test = y[100:]
```

```python
from operator import xor
from sklearn.decomposition import PCA

# we create a PCA object
pca = PCA(0.95)

# we fit the dataset in X
pca.fit(X_train)
pca.fit(X_test)

# we extract the principal components directions
print('first principal component direction:', pca.components_[:,0])

# and the corresponding explained variance
explained = pca.explained_variance_
print('explained variance:', explained)

T = pca.transform(X_train)
T_dot = pca.transform(X_test)

print ("Modified data ", T[0])
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import RepeatedStratifiedKFold

logistic = LogisticRegression(multi_class='ovr', solver='lbfgs')

logistic.fit(T, y_train)

y_hat = logistic.predict(T_dot)

vali_accuracy = sum(y_hat == y_test) / len(y_test)
print('Validation accuracy:', vali_accuracy)
```
